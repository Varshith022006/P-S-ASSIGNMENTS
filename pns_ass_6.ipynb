```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean = 30
std_dev = 7
data = np.random.normal(mean, std_dev, 1000)
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='black', edgecolor='yellow')
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r', label='PDF')
plt.title('Normal Distribution with Mean = 30, Std Dev = 7')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean_rhr = 50
std_dev_rhr = 10
prob_greater_30 = 1 - norm.cdf(30, mean_rhr, std_dev_rhr)
prob_less_30 = norm.cdf(30, mean_rhr, std_dev_rhr)
prob_between_30_50 = norm.cdf(50, mean_rhr, std_dev_rhr) - norm.cdf(30, mean_rhr, std_dev_rhr)
x = np.linspace(mean_rhr - 4*std_dev_rhr, mean_rhr + 4*std_dev_rhr, 100)
pdf_rhr = norm.pdf(x, mean_rhr, std_dev_rhr)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_rhr, 'b', label='PDF of RHR')
plt.fill_between(x, pdf_rhr, where=(x >= 30), color='lightgreen', alpha=0.5, label=f'P(RHR  30) = {prob_greater_30:.3f}')
plt.fill_between(x, pdf_rhr, where=(x <= 30), color='lightcoral', alpha=0.5, label=f'P(RHR  30) = {prob_less_30:.3f}')
plt.fill_between(x, pdf_rhr, where=((x >= 30) & (x <= 50)), color='lightskyblue', alpha=0.5, label=f'P(30  RHR  50) = {prob_between_30_50:.3f}')
plt.title('Probability Distribution of Resting Heart Rate (RHR)')
plt.xlabel('Resting Heart Rate (bpm)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
print(f"P(RHR  30 bpm) = {prob_greater_30:.3f}")
print(f"P(RHR  30 bpm) = {prob_less_30:.3f}")
print(f"P(30 bpm  RHR  50 bpm) = {prob_between_30_50:.3f}")
```
