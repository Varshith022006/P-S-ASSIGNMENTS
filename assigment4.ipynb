```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/car_data_250.csv')
plt.figure(figsize=(8, 6))
plt.hist(df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price (in USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Horsepower', y='Mileage', hue='Fuel Type', palette='Set1')
plt.title('Horsepower vs. Mileage by Fuel Type')
plt.xlabel('Horsepower')
plt.ylabel('Mileage (mpg)')
plt.legend(title='Fuel Type')
plt.grid(True)
plt.show()
brand_counts = df['Brand'].value_counts()
plt.figure(figsize=(8, 6))
brand_counts.plot(kind='bar', color='lightcoral')
plt.title('Number of Cars per Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
fuel_type_by_year = df.groupby(['Year', 'Fuel Type']).size().unstack().fillna(0)
fuel_type_by_year.plot(kind='bar', stacked=True, figsize=(10, 6), cmap='viridis')
plt.title('Fuel Type Preference by Manufacturing Year')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.legend(title='Fuel Type')
plt.grid(axis='y')
plt.show()
price_range = pd.cut(df['Price'], bins=[0, 20000, 40000, 60000, 80000, 100000, float('inf')])
most_common_price_range = price_range.value_counts().idxmax()
print(f'Most common price range: {most_common_price_range}')
correlation = df[['Horsepower', 'Mileage']].corr().iloc[0, 1]
print(f'Correlation between Horsepower and Mileage: {correlation:.2f}')
if correlation < 0:
    print("There is a negative correlation between horsepower and mileage (higher horsepower generally leads to lower mileage).")
else:
    print("There is a positive correlation between horsepower and mileage.")
most_popular_brand = brand_counts.idxmax()
print(f'The most popular car brand is: {most_popular_brand}')
```
